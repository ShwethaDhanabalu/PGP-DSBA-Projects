<b>2.2 Calculate Returns for all stocks with inference

```python
data_stock_prices
```

```python
data_stock_prices.drop('Date',axis =1,inplace = True)
```

```python
data_stock_prices
```

```python
returns = np.log(data_stock_prices).diff(axis = 0, periods = 1) 
returns
```

```python
returns.info()
```

```python
returns.describe().T
```

<b>2.3 Calculate Stock Means and Standard Deviation for all stocks with inference

```python
returns_mean = returns.mean(axis = 0)
returns_mean
```

```python
returns_std = returns.std(axis = 0)
returns_std
```

```python
stock_measures = pd.DataFrame({'Average':returns_mean, 'Volatility': returns_std})
stock_measures
```

<b>2.4 Draw a plot of Stock Means vs Standard Deviation and state your inference

```python
plt.figure(figsize = (10, 7))
plot = sns.scatterplot(stock_measures['Volatility'], stock_measures['Average'],hue=stock_measures.index,s=200)
plt.title('Stock price average mean vs Standard deviation')
plt.show()
```
